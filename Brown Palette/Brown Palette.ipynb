```python
import numpy as np
import matplotlib.pyplot as plt

# Define ranges for R, G, B values
r_values = np.arange(100, 161, 10)  # R: 130 +/- 30
g_values = np.arange(45, 106, 10)   # G: 75 +/- 30
b_values = np.arange(0, 31, 10)     # B: 0 to 30

# Generate all combinations of R, G, B within the specified ranges
brown_palette = np.array(np.meshgrid(r_values, g_values, b_values)).T.reshape(-1, 3)

# Convert values to scale [0, 1] for displaying with matplotlib
brown_palette_normalized = brown_palette / 255.0

# Display the color palette
fig, ax = plt.subplots(1, figsize=(10, 5))
ax.imshow([brown_palette_normalized], aspect='auto')
ax.set_title('Different shades of Brown images ')
ax.axis('off')  # Hide the axis
plt.show()
```
