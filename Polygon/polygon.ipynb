```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_polygon_vertices(n, radius=1):
    vertices = []
    angle_step = 2 * np.pi / n  # Angle between vertices (in radians)
    
    for i in range(n):
        theta = i * angle_step  # Angle for the i-th vertex
        x = radius * np.cos(theta)  # x coordinate
        y = radius * np.sin(theta)  # y coordinate
        vertices.append((x, y))
    
    return vertices
```

```python
# Example: Generate vertices for polygons with different sides (n)
n_sides = [3, 4, 5, 6]  # Triangle, Square, Pentagon, Hexagon
```

```python
for n in n_sides:
    vertices = generate_polygon_vertices(n)
    print(f"\nVertices of a {n}-sided polygon:")
    for i, (x, y) in enumerate(vertices):
        print(f"Vertex {i+1}: ({x:.2f}, {y:.2f})")
```

```python
# Optional: Plot the polygons
plt.figure()
for n in n_sides:
    vertices = generate_polygon_vertices(n)
    vertices.append(vertices[0])  # Close the polygon by adding the first vertex at the end
    polygon = np.array(vertices)
    
    plt.plot(polygon[:, 0], polygon[:, 1], 'o-', label=f'{n}-sided polygon')
    
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Polygons with different number of sides')
plt.legend()
plt.grid(True)
plt.show()
```
